## Vortex Lattice Method 

```python
angulo = float(input("Ingrese ángulo de ataque: "))
alpha= math.radians(angulo)
vel_inf = 24.32 #m/s
```

### Variables

```python
x = []
y = []
x_1 = []
y_1 = []
x_2 = []
y_2 = []

for i in range(11):
    xi = 2*i/100
    x_1.append(xi)
    yi = round((-0.1*xi+0.045),2)
    y_1.append(yi)
    

for i in range(11):
    x1i = 2*i
    x_2.append(x1i)
    y1i = round((-0.1*x1i+(0.045/2)),2)
    y_2.append(y1i)

    
for i in range(10):
    xii = (x_1[i]+x_1[i+1])/2  
    x.append(xii)
    yii = round(((y_1[i]+y_2[i+1])/2 ),2)
    y.append(yii)
    
for i in range(10):
    x.append(x[i])
    yie = round(((y_2[i]+y_2[i+1])/4 ),2)
    y.append(yie)
    
print("x = " , x)
print("")
print("y = " , y)

h = 0.002
print("")
print("h = ", h)

k = 0.01
den = 1.194
```

```python
import math
A_ij = []
A_filas = []
for i in range(20):
    A_filas = []
    for j in range(20):
        R1 = math.sqrt(((x[i]+h)**2)+((y[j]+k)**2))
        R2 = math.sqrt(((x[i]-h)**2)+((y[j]-k)**2))
        a_boundvortex = ((1)/((x[i]+h)*(y[j]-k)-(x[i]-h)*(y[j]+k))) *(((2*h*(x[i]+h)+2*k*(y[j]+k))/(R1))-((2*h*(x[i]-h)+2*k*(y[j]-k))/(R2)))
        a_lefttrailing = (-1/(y[i]+k))*(1+((x[i]+h)/R1))
        a_rigth_trailing = (1/(y[i]-k))*(1+((x[i]-h)/R2))
        a_ij = round(((1/4*math.pi)*(a_boundvortex+a_lefttrailing+a_rigth_trailing)),4)
        A_filas.append(a_ij)
    A_ij.append(A_filas)
        
        
for i in A_ij:
    for j in i:
        print(j, end="\t")
        
    print("\n", end="")
```

```python
inversa_Aij = []
    
from numpy import matrix
from scipy import linalg

inversa_Aij = linalg.inv(A_ij)
print(inversa_Aij)
```

```python
sin_theta1 = alpha-0.0045
print(sin_theta1)
sin_theta2 = alpha+0.000664970445

vector_x = []
for i in range(20):
    if i <= 9:
        dato1 = round((-vel_inf*sin_theta1),3)
        vector_x.append(dato1)
    else:
        dato2 = round((-vel_inf*sin_theta2),3)
        vector_x.append(dato2)
        
print(vector_x)
```

```python
import numpy as np

circulaciones = np.dot(inversa_Aij,vector_x)
print(circulaciones)
```

```python
import numpy as np

circulaciones2 = np.dot(inversa_Aij,vector_x2)
print(circulaciones2)
```
